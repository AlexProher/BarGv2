```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
from scipy import interpolate

way = '/Users/lubimyj/Git/experiments/Plansee-Tungstan-W_cover/'
material = ['Denal920','DX2']
specimen = ['1/','2/','3/','4/','5/','6/']
subFold = '/IRCamera/Specimen'
testFold = 'test/EXP #1/'
dataFold = '/FLT/'
files = ['Parameters.txt','incid.FLT','trans.FLT']

def find_max(data):
    ind_max = data.idxmax()
    value = data[ind_max-10:ind_max+10].mean()
    return value

    
def my_interpolation(data, len_aim):
    xnew = np.arange(0,len(data.dropna()),len(data.dropna())/len_aim)
    xnew[-1] = xnew[-1]-1
    x = np.arange(len(data.dropna()))
    y = np.array(data.dropna())
    f = interpolate.interp1d(x, y)
    ynew = f(xnew)
    return (ynew)
```

Дальше будем наводить статистику

Сначала нужно сделаль все массивы одинаковой длины, для того чтобы их можно было осреднять

```python
stress_denial_all = pd.DataFrame()
strain_denial_all = pd.DataFrame()
for ind in [0,1,2,3,4,5]:
    stress_denial_all['samp_'+str(ind+1)] = data['specimen_'+specimen[ind][0]]['true_stress']
    strain_denial_all['samp_'+str(ind+1)] = data['specimen_'+specimen[ind][0]]['true_strain']

stress_dx_all = pd.DataFrame()
strain_dx_all = pd.DataFrame()
for ind in [0,1,2,3,4,5]:
    stress_dx_all['samp_'+str(ind+1)] = data_dx['specimen_'+specimen[ind][0]]['true_stress']
    strain_dx_all['samp_'+str(ind+1)] = data_dx['specimen_'+specimen[ind][0]]['true_strain']



stress_denial_all['samp_5'] = my_interpolation(stress_denial_all['samp_5'], len(stress_denial_all['samp_1']))
strain_denial_all['samp_5'] = my_interpolation(strain_denial_all['samp_5'], len(strain_denial_all['samp_1']))

stress_dx_all['samp_6'] = my_interpolation(stress_dx_all['samp_6'], len(stress_dx_all['samp_2']))
strain_dx_all['samp_6'] = my_interpolation(strain_dx_all['samp_6'], len(strain_dx_all['samp_2']))

stress_dx_all['samp_4'] = my_interpolation(stress_dx_all['samp_4'], len(stress_dx_all['samp_2']))
strain_dx_all['samp_4'] = my_interpolation(strain_dx_all['samp_4'], len(strain_dx_all['samp_2']))

stress_dx_all['samp_5'] = my_interpolation(stress_dx_all['samp_5'], len(stress_dx_all['samp_2']))
strain_dx_all['samp_5'] = my_interpolation(strain_dx_all['samp_5'], len(strain_dx_all['samp_2']))

stress_denial_all['mean'] = stress_denial_all.mean(axis=1) 
strain_denial_all['mean'] = strain_denial_all.mean(axis=1) 
strain_dx_all['mean'] = strain_dx_all.mean(axis=1) 
stress_dx_all['mean'] = stress_dx_all.mean(axis=1) 

stress_denial_all['std'] = stress_denial_all.std(axis=1) 
strain_denial_all['std'] = strain_denial_all.std(axis=1) 
strain_dx_all['std'] = strain_dx_all.std(axis=1) 
stress_dx_all['std'] = stress_dx_all.std(axis=1) 
```

```python
fig1 = go.Figure()
fig1.update_layout(template = 'plotly_white', width=800, height=600,
                  title = f"Material - {material[0]} True stress-strain curves")
for ind in stress_denial_all.columns[:-2]:
    
    fig1.add_trace(go.Scatter(x = strain_denial_all[ind],
                             y = stress_denial_all[ind],
                             name = ind))

#fig1.add_trace(go.Scatter(x = strain_denial_all['mean'][::10], 
#                            y = stress_denial_all['mean'][::10],
#                            error_y = dict(array = stress_denial_all['std'][::10]),
#                            name = 'Mean Value'))
fig1.update_yaxes(title = 'Stress, MPa')
fig1.update_xaxes(title = 'Strain')
fig1.show()
```
