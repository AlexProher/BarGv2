```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
from scipy import interpolate

way = '/Users/lubimyj/Git/experiments/Plansee-Tungstan-W/'
material = ['Denal920','DX2']
specimen = ['Specimen1','Specimen2','Specimen3','Specimen4','Specimen5','Specimen6']
subFold = 'results_'

files = ['Parameters.txt','incid.FLT','trans.FLT']

def find_max(data):
    ind_max = data.idxmax()
    value = data[ind_max-10:ind_max+10].mean()
    return value

    
def my_interpolation(data, len_aim):
    xnew = np.arange(0,len(data.dropna()),len(data.dropna())/len_aim)
    xnew[-1] = xnew[-1]-1
    x = np.arange(len(data.dropna()))
    y = np.array(data.dropna())
    f = interpolate.interp1d(x, y)
    ynew = f(xnew)
    return (ynew)
```

```python
way_to_file = way + subFold + material[1] +'/'+ specimen[0] + '/'
res = pd.read_csv(way_to_file + 'Stress-Strain True.csv')
```

```python
data = {item: 
        pd.read_csv(way + subFold + material[0] +'/'+ item + '/' + 'Stress-Strain True.csv')
        for item in specimen}
```

```python
data_dx = {item: 
        pd.read_csv(way + subFold + material[1] +'/'+ item + '/' + 'Stress-Strain True.csv')
        for item in specimen}
```

```python
def validate_sign(stress, strain):
    return(stress.max(), stress.mean(), strain.max(), strain.mean(), stress.sum(), strain.sum())
```

```python
for item in data.keys():
    print(validate_sign(data[item][' Stress'],data[item]['# Strain']))
```

```python
stress_denial = pd.DataFrame({'stress_val':[find_max(data[item][' Stress']) for item in specimen],
                            'max_strain': [data[item]['# Strain'].max() for item in specimen]},
                            index=specimen)
stress_dx = pd.DataFrame({'stress_val':[find_max(data_dx[item][' Stress']) for item in specimen],
                            'max_strain': [data_dx[item]['# Strain'].max() for item in specimen]},
                            index=specimen)
```

```python
stress_denial.loc['mean'] = stress_denial.mean()
stress_denial.loc['std'] = stress_denial.std()

stress_dx.loc['mean'] = stress_dx.mean()
stress_dx.loc['std'] = stress_dx.std()

stress_dx
```
