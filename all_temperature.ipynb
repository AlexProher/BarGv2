```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
way = '/Users/lubimyj/Git/experiments/Plansee-Tungstan-W/'
materials = ['Denal920','DX2']
specimens = ['1','2','3','4','5','6']
to_IR_data = '/Specimen'
to_raw_data = '/test/Original WFT Files/'
IR_index = '_IR/'
subFold = '/IRCamera/Specimen'
testFold = 'test/EXP #1/'
dataFold = '/FLT/'
title = 'Tr_0.9'
max_data = pd.DataFrame() 

material = materials[0]
print (f'material {material}')

fig1 = go.Figure()
fig1.update_layout(template = 'plotly_white', width=800, height=600,
                        title = f'{material} {title}')
for specimen in specimens:
    
    thermo_data_path = way + material + subFold + specimen + to_IR_data + specimen + IR_index

    thermo_data = pd.read_csv(thermo_data_path +'Temperature_Spec_'+specimen+'Max_Tr.csv', sep = '\t' )
    fig1.add_trace(go.Scatter(x = thermo_data.time, y = thermo_data[title], name = f'specimen_{specimen}'))
    fig1.update_xaxes(title_text="Time, s")

    fig1.update_yaxes(title_text="Temperature rise, C")
    max_data[f'specimen_{specimen}'] = thermo_data[title]
max_data['time'] = thermo_data['time']
fig1.show()
#fig1.write_html(way + material+ '/' + f'All_IR_data{material}_{title}.html')
```
